#converted 0 and 1 values of senior citizen to yes\no to make it easier to better understand

```python
def conv(value):        # convert 0 to 1 in no and yes to better understantnd 
    if value == 1:
        return "yes"
    else:
        return "No"
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv) 
```

```python
# coluld not convert string to float, that why create group by through u can count churn 
```

```python
plt.figure (figsize = (4,4))
gp = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gp["Churn"],labels  = gp.index,autopct = "%1.2f%%")
plt.title("percentage of churn customer", fontsize = 10)
plt.show()
```

```python
df.columns.values
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))  # 3 rows, 3 columns
axes = axes.flatten()  # Flatten axes to easily iterate over them

# Loop through the columns and axes to create countplots
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i] , hue = "Churn")
    axes[i].set_title(col)  # Title for each subplot
    axes[i].set_xlabel('')  # Optional: remove x-label for better spacing
    axes[i].set_ylabel('Count')  # Set y-label as Count

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

#This data shows customer churn rates across various services for a telecom company. Generally, customers with "No" for optional services like 
OnlineSecurity, TechSupport, and DeviceProtection tend to have higher churn rates compared to those who use these services. Fiber optic internet 
users also show a higher churn rate than DSL users. Additionally, customers with "Yes" for PhoneService and Streaming services have a relatively 
lower churn rate.
